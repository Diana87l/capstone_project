- seleziono alcune colonne per creare un set customizzato e piÃ¹ piccolo rispetto quello iniziale

```python
df=df[['country','year','energy_per_gdp', 'energy_per_capita','gdp', 'population']]
```

- scelgo di analizzare i Paesi dell' OPEC e alcuni extra ( Stati Uniti, Giappone, India, Canada, Cina, Russia)

```python
df=df[df['year']>=2000]
countries=['Algeria','Angola', 'Congo', 'Equatorial Guinea', 'Gabon', 'Iran', 'Iraq', 'Kuwait','Lybia', 'Nigeria', 'Saudi Arabia',
           'United Arab Emirates', 'Venezuela','United States','Japan','India', 'Canada', 'China', 'Russia']
filtro_countries = df.loc[:, "country"].isin(countries)
df = df.loc[filtro_countries]
```
