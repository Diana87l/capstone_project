- a questo punto seleziono i Paesi e gli attributi/colonne di interesse, scelgo di inizializzare l'analisi a partire dagli anni 2000.

```python
df=df[['country','year','biofuel_electricity','coal_electricity','fossil_electricity','gas_electricity','hydro_electricity',
        'oil_electricity', 'other_renewable_electricity','other_renewable_exc_biofuel_electricity','renewables_electricity',
        'solar_electricity','wind_electricity','low_carbon_electricity', 'nuclear_electricity']]
```

```python
df=df[df['year']>=2000]
countries=['Algeria','Angola', 'Congo', 'Equatorial Guinea', 'Gabon', 'Iran', 'Iraq', 'Kuwait','Lybia', 'Nigeria', 'Saudi Arabia',
           'United Arab Emirates', 'Venezuela','United States','Japan','India', 'Canada', 'China', 'Russia']
filtro_countries = df.loc[:, "country"].isin(countries)
df = df.loc[filtro_countries]
```
